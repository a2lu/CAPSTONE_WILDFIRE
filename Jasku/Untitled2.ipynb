```python
crs = 'EPSG:4326'

# CA counties
counties = gpd.read_file("/content/drive/Shareddrives/Capstone 2022/Oscar/data/CA_Counties/CA_Counties_TIGER2016.shp")
counties.to_crs(crs, inplace=True)

# San Francisco lower bound
SF = counties[counties["NAME"] == "San Francisco"]
lowerBound = SF["geometry"].bounds["maxy"].values[0]

norCalCounties = counties[counties.bounds.apply(lambda x: x[3] > lowerBound, axis=1)]
norCalCounties
```

```python
alt.layer(alt.Chart(norCalCounties
            ).mark_geoshape(fill="#E6E6E6", stroke="black"
            ).encode(tooltip=["NAME"]))
```
